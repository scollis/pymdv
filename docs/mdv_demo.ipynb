Load in a NCAR Meterological Data Volume (MDV) file containing CSAPR data into a pymdv object

```python
filename='/data/sample_sapr_data/235545.mdv'#Replace with the location of your MDV file, See tutorial on fetching ARM data 
my_mdv_object=py_mdv.read_mdv(filename, debug=True)
```

my_mdv_object is now a pymdv object containing the metadata and some methods.. At init the object does not populate the data into the object to save time (in case the user only wants certain moments..). Lets take a look at what is in the object. 

```python
print dir(my_mdv_object)
print "Radar metadata"
print my_mdv_object.radar_info.keys()
```

Ok, so the objects in capitals are chunks of metadata read straight from the MDV file.. you can generally ignore those, the objects in lower caps are methods and data.. Lets have a look at things describing the geometry: range_km, el_deg and az_deg.

```python
print my_mdv_object.scan_type
print my_mdv_object.range_km[0], " ", my_mdv_object.range_km[1], " ", my_mdv_object.range_km[-1]
print len(my_mdv_object.el_deg)
print my_mdv_object.el_deg
```

Ok, lets get some data! Lets see what fields we have.

```python
print my_mdv_object.fields
```

Note that these are as described (poorly) in the MDV file. The ARM datastream data base has more descriptive names. These also refer to the indexing used in calling the method read_a_field. Lets grab some reflectivity data. 


```python
print my_mdv_object.fields.index('DBZ_F')
my_z=my_mdv_object.read_a_field(my_mdv_object.fields.index('DBZ_F'))
```

this does tow things, saves it to my_z and also to my_mdv_object.DBZ_F, lets inspect!

```python
print my_z.shape
print my_z[0,180,0:10]
```

```python
figure(figsize=[10,4])
plot(my_mdv_object.range_km, my_z[0,180,:])
xlabel('Range (km)')
ylabel('Reflectivity (dBZ)')
title(my_mdv_object.radar_info['radar_name']+" Reflectivity")

```

Or, since we have a nice polarimetric data lets look at the propigation phase and the differential reflectivity!

```python
my_zdr=my_mdv_object.read_a_field(my_mdv_object.fields.index('ZDR_F'))
my_phdp=my_mdv_object.read_a_field(my_mdv_object.fields.index('PHIDP_F'))
figure(figsize=[12,8])
subplot(2,1,1)
plot(my_mdv_object.range_km, my_zdr[0,180,:])
xlabel('Range (km)')
ylabel('Differential Reflectivity (dB)')
title(my_mdv_object.radar_info['radar_name']+" Differential Reflectivity")
subplot(2,1,2)
plot(my_mdv_object.range_km, my_phdp[0,180,:])
xlabel('Range (km)')
ylabel('Propigation Phase (Degrees)')
title(my_mdv_object.radar_info['radar_name']+" Propigation Phase")

```

Lots of interesting physics here! Note the negative $Z_{DR}$, this is due to differential attenuation. We are working towards a correction for this. Also in the $\phi_{DP}$ note the negative gradient at ~75km.. this is caused by either non-uniform beam filling or differential phase on backscatter ($\delta_{DP}$) or a combination of both.. again work in ARM is active in this area!

Oh and the sudden jump at the end, that is an injected test signal.. whenever you do quantitative work be sure to dock 10 or so gates off the end.

So lets look at a two dimensional representation, or a PPI, this is what we are commonly used to seeing on the web (eg at [The ARM radar website](http://radar.arm.gov/ "Radar website")). Even better, the pymdv class has already calculated x,y,z coordinates for the gates assuming a $\frac{4}{3}$ earth radius propoigatuon assumption (see Doviac and Zrnic, 1992):


$R_v=\frac{4r_e}{3}$

$z=\sqrt{r^2 + R_v^2+ 2rR_v\sin{}(\frac{\phi\pi{}}{180})}-R_v$

$s=R_v\arcsin(r\cos(\phi\pi/180)/(R_v+z))$

$x=s\sin(\theta\pi/180)$

$y=s\cos(\theta\pi/180)$

Ok, so those with a keen eye can pick up there are some funny looking returns close to the radar to the east. These are what we call second trip returns. Lets zoom in.

```python
tilt=1
figure(figsize=[12,9])
pcolormesh(my_mdv_object.carts['x'][tilt,:,:]/1000.0, 
           my_mdv_object.carts['y'][tilt,:,:]/1000.0, 
           my_mdv_object.read_a_field(my_mdv_object.fields.index('DBZ_F'))[tilt,:,:], vmin=-8, vmax=72)
xlim([-10,120])
ylim([-60,60])
cb=colorbar()
cb.set_label('Eq Refl Fac (dBz)')
```

Now the nice thing about the magnetron transmitter in the ARM C-SAPR is that the phase is randomized from pulse to pulse, so although second trips are generating a significant return, lets look at thier Normalized Coherent Power or NCP: 

```python
tilt=1
figure(figsize=[12,9])
pcolormesh(my_mdv_object.carts['x'][tilt,:,:]/1000.0, 
           my_mdv_object.carts['y'][tilt,:,:]/1000.0, 
           my_mdv_object.read_a_field(my_mdv_object.fields.index('NCP_F'))[tilt,:,:], vmin=0, vmax=1)
xlim([-10,120])
ylim([-60,60])
cb=colorbar()
cb.set_label('Normalized Coherent Power (frac)')
```

So we can use Numpy's masked array class to do some nice thresholding on NCP

```python
from numpy import ma
my_ncp=my_mdv_object.read_a_field(my_mdv_object.fields.index('NCP_F'))
my_masked_z=ma.masked_where(my_ncp<0.4, my_z)
tilt=1
figure(figsize=[12,9])
pcolormesh(my_mdv_object.carts['x'][tilt,:,:]/1000.0, 
           my_mdv_object.carts['y'][tilt,:,:]/1000.0, 
           my_masked_z[tilt,:,:], vmin=-8, vmax=72)
xlim([-10,120])
ylim([-60,60])
cb=colorbar()
cb.set_label('Eq Refl Fac (dBz)')
```

So not perfect, but can nicely filter out most of the second trips... And there are all kinds of games we can play with texture mapping and using some of the polarimetric moments to identify clutter etc.. **A warning for new players:** Areas of large spectral widths and high returns, ie convective cores, can have very high NCP, filtering second trip returns robustly is not a simple process.

Ok, what can we do quantitively without mapping to a cartesian grid (the subject of a future tutorial)? Well lets partition the reflectivity by height... First lets flatten (create a 1d array) for the z coordinates and values. Oh and lets doc off the test signal at the end...

```python
doc=20
refl_flat=my_z[:,:,0:-1*doc].flatten()
z_flat=my_mdv_object.carts['z'][:,:,0:-1*doc].flatten()
```

So now we have two arrays, an array of reflectivities and an array of thier corresponding heights above the radar. we can neatly partition this into height bins.. 

```python

my_heights=linspace(0,17,18*2)*1000.0
d_height=my_heights[1]-my_heights[0]
height_part_list=[]
for height in my_heights:
    lower=height-d_height/2
    upper=height+d_height/2
    print "searching for values between ", lower/1000.0, "km and ", upper/1000.0,"km"
    lesser=where(z_flat > lower)[0]
    greater=where(z_flat < upper)[0]
    between=list(set(greater)&set(lesser))
    height_part_list.append(refl_flat[between])
```

So now wa have a list of arrays which contain height partioned values of reflectivity, so lets look at a height dependent histogram or Contoured Frequency by Altitude Diagram (CFAD)

```python
import scipy.stats
```

```python
cfad=zeros([len(my_heights), 60], dtype=float)
for i in range(len(my_heights)):
    this_level=scipy.stats.histogram(height_part_list[i], 60, (0,60))
    cfad[i, :]=this_level[0]/len(height_part_list[i])
```

Now lets plot it!

```python
figure(figsize=[8,5])
pcolormesh(linspace(0,60,60), my_heights/1000.0, cfad)
colorbar()
```

Not to enlightening as so many of the values are around 0dBz.. Clearly we need to do a conditional PDF to tease out the relevant meterological returns.. Ahh but this is a polarimetric radar so we have the correlation cooeficent $\rho_{hv}$ 

```python
filtered_zs=[]
rhv_flat=my_mdv_object.read_a_field(my_mdv_object.fields.index('RHOHV_F'))[:,:,0:-1*doc].flatten()
good_rhvs=where(rhv_flat > 0.5)[0] #Meteorological values are defined as being correlated greater than 0.5
good_rfl=refl_flat[good_rhvs]
good_z=z_flat[good_rhvs]
for i in range(len(my_heights)):
    lower=my_heights[i]-d_height/2
    upper=my_heights[i]+d_height/2
    print "searching for values between ", lower/1000.0, "km and ", upper/1000.0,"km"
    lesser=where(good_z > lower)[0]
    greater=where(good_z < upper)[0]
    between=list(set(greater)&set(lesser))
    filtered_zs.append(good_rfl[between])
```

```python
cfad=zeros([len(my_heights), 60], dtype=float)
for i in range(len(my_heights)):
    this_level=scipy.stats.histogram(filtered_zs[i], 60, (0,60))
    if len(filtered_zs[i]) >300:
        cfad[i, :]=this_level[0]/len(filtered_zs[i])
    else:
        cfad[i, :]=this_level[0]*0.0 #filter out sparse results
```

```python
figure(figsize=[10,5])
pcolormesh(linspace(0,60,60), my_heights/1000.0, cfad*100.0, vmin=0, vmax=15, cmap='gist_ncar')
xlabel('Reflectivity (dBZ)')
ylabel('Height (km)')
title('Contoured frequency by height diagram')
cb=colorbar()
cb.set_label('Occurance (%)')
```

Interesting! You can clearly see the increase in Z towards the freezing level where you get a large spread and mean Z and then a reduction towards the surface. The returns at ~16km are interesting, overshooting tops?
